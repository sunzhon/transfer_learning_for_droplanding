```python
import time
import re
import os

rename_file="linux_subjects_data.log"
if(os.path.exists(rename_file)):
    os.remove(rename_file)


for idx, line in enumerate(open("./subjects_data.log")): 
    ss=re.search('test_data',line)
    if(ss!=None):
        print(line)
        #print(ss)
        index=ss.span(0)[0]
        #print(index)
        short_line=line[index:-1]
        print("short_line:",short_line)
        fine_line=re.sub(r"\\",'/',short_line)
        line="\'"+re.sub("test_data","All_data",fine_line)
        print("modify line:",line)
        print("----------------")
    with open(rename_file,"a") as mon:
        mon.write(line)  
    

```
